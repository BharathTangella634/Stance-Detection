```python
duplicate_rows = (
    df_all.groupby(['Tweet', 'Target 1'])
    .filter(lambda x: len(x) > 1)
)
print(len(duplicate_rows)) # No of samples to consider
duplicate_rows  # text, target pair repeated more than once
```

```python
result = []
grouped = duplicate_rows.groupby(['Tweet', 'Target 1'])

for (tweet, target), group in grouped:
    stance_counts = [
        (group['Stance 1'] == 'AGAINST').sum(),  # Count AGAINST
        (group['Stance 1'] == 'FAVOR').sum(),    # Count FAVOR
        (group['Stance 1'] == 'NONE').sum()      # Count NONE
    ]
    result.append([[tweet, target], stance_counts])

# Print the result
print(result[0]) # [[text, target], [against_count, favor_count, none_count]]
print(len(result))
```

```python
# non-problematic samples with duplicates
count = 0
total_samples = 0
for i in range(len(result)):
  if(result[i][1].count(0) >= 2):
    # print(result[i][1])  # uncomment this to view clearly
    count += 1
    total_samples += sum(result[i][1])
print("Total samples :", total_samples)
print("Count of unique samples :", count)
```

```python
# 1897 - 857 = 1040 duplicates has to be removed.
df_all = df_all.drop_duplicates()
print(len(df_all))
df_all.head()
```

```python
# problematic samples and also with duplicates
count = 0
total_samples = 0
for i in range(len(result)):
  if(result[i][1].count(0) < 2):
    # print(result[i][1])  # uncomment this to view clearly
    count += 1
    total_samples += sum(result[i][1])
print("Total samples :", total_samples)
print("Count of unique samples :", count)
```

```python
# from 1132 we have to take 467 unique text-target pairs and predict using gemini-api.
df_pred = pd.read_csv("/content/predicted_stances_467.csv")
print(len(df_pred))
df_pred.tail()
```

```python
# Now remove those 1132 samples from that original data.
for sample, cts in result:
  if(cts.count(0) < 2):
    df_all = df_all[~((df_all["Tweet"] == sample[0]) & (df_all["Target 1"] == sample[1]))]
print(len(df_all))
print(df_all.tail())
```
